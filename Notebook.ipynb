```python
y_class=df['encoded_decision']
y_reg=df["css_creditscorefinal"]
x=df.drop(['encoded_decision','css_creditscorefinal'],axis=1)
```

```python
x= pd.get_dummies(x)
```

```python
import numpy as np
from sklearn.decomposition import PCA


pca = PCA(n_components=2)
# Check DataFrame types
print(x.dtypes)

# Exclude non-numeric columns
numeric_columns = x.select_dtypes(exclude=['datetime64', 'object'])

# Apply PCA to numeric columns
pca = PCA(n_components=2)
x_pca = pca.fit_transform(numeric_columns)


explained_variance_ratio = pca.explained_variance_ratio_
print("Explained Variance Ratio:", explained_variance_ratio)

principal_components = pca.components_
print("Principal Components (Eigenvectors):", principal_components)

singular_values = pca.singular_values_
print("Singular Values:", singular_values)
```

```python
import matplotlib.pyplot as plt

# Visualize explained variance
cumulative_explained_variance = np.cumsum(explained_variance_ratio)
plt.plot(range(1, len(cumulative_explained_variance) + 1), cumulative_explained_variance, marker='o')
plt.xlabel('Number of Principal Components')
plt.ylabel('Cumulative Explained Variance')
plt.show()
```

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA


# Standardize the numeric columns
scaler = StandardScaler()
X_standardized = scaler.fit_transform(numeric_columns)

# Apply PCA to the standardized data with 6 components
n_components = 6
pca = PCA(n_components=n_components)

# Fit and transform the data in one step
x_new = pca.fit_transform(X_standardized)
```

```python
from sklearn.model_selection import train_test_split
x_train_r, x_test_r, y_train_r, y_test_r = train_test_split(x_new, y_reg, test_size=0.2, random_state=42)
x_train_c, x_test_c, y_train_c, y_test_c = train_test_split(x_new, y_class, test_size=0.2, random_state=42)
```

```python
import pandas as pd
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

# Initialize XGBoost classifier
xgb_classifier = XGBClassifier()

# Fit the model on the training data
xgb_classifier.fit(x_train_c, y_train_c)

# Make predictions on the test data
y_pred = xgb_classifier.predict(x_test_c)

# Evaluate the model
accuracy = accuracy_score(y_test_c, y_pred)
conf_matrix = confusion_matrix(y_test_c, y_pred)
class_report = classification_report(y_test_c, y_pred)

# Display the evaluation metrics
print("Accuracy:", accuracy)
print("Confusion Matrix:\n", conf_matrix)
print("Classification Report:\n", class_report)
```

```python
from sklearn.svm import SVC

# Initialize SVM classifier
svm_classifier = SVC()

# Fit the model on the training data
svm_classifier.fit(x_train_c, y_train_c)

# Make predictions on the test data
y_pred_svm = svm_classifier.predict(x_test_c)

# Evaluate the model
accuracy_svm = accuracy_score(y_test_c, y_pred_svm)
conf_matrix_svm = confusion_matrix(y_test_c, y_pred_svm)
class_report_svm = classification_report(y_test_c, y_pred_svm)

# Display the evaluation metrics for SVM
print("SVM Accuracy:", accuracy_svm)
print("SVM Confusion Matrix:\n", conf_matrix_svm)
print("SVM Classification Report:\n", class_report_svm)
```

```python
from sklearn.ensemble import RandomForestClassifier

# Initialize Random Forest classifier
rf_classifier = RandomForestClassifier()

# Fit the model on the training data
rf_classifier.fit(x_train_c, y_train_c)

# Make predictions on the test data
y_pred_rf = rf_classifier.predict(x_test_c)

# Evaluate the model
accuracy_rf = accuracy_score(y_test_c, y_pred_rf)
conf_matrix_rf = confusion_matrix(y_test_c, y_pred_rf)
class_report_rf = classification_report(y_test_c, y_pred_rf)

# Display the evaluation metrics for Random Forest
print("Random Forest Accuracy:", accuracy_rf)
print("Random Forest Confusion Matrix:\n", conf_matrix_rf)
print("Random Forest Classification Report:\n", class_report_rf)
```

```python
from sklearn.neighbors import KNeighborsClassifier

# Initialize KNN classifier
knn_classifier = KNeighborsClassifier()

# Fit the model on the training data
knn_classifier.fit(x_train_c, y_train_c)

# Make predictions on the test data
y_pred_knn = knn_classifier.predict(x_test_c)

# Evaluate the model
accuracy_knn = accuracy_score(y_test_c, y_pred_knn)
conf_matrix_knn = confusion_matrix(y_test_c, y_pred_knn)
class_report_knn = classification_report(y_test_c, y_pred_knn)

# Display the evaluation metrics for KNN
print("KNN Accuracy:", accuracy_knn)
print("KNN Confusion Matrix:\n", conf_matrix_knn)
print("KNN Classification Report:\n", class_report_knn)
```

```python
from sklearn.naive_bayes import GaussianNB

# Initialize Gaussian Naive Bayes classifier
gnb_classifier = GaussianNB()

# Fit the model on the training data
gnb_classifier.fit(x_train_c, y_train_c)

# Make predictions on the test data
y_pred_gnb = gnb_classifier.predict(x_test_c)

# Evaluate the model
accuracy_gnb = accuracy_score(y_test_c, y_pred_gnb)
conf_matrix_gnb = confusion_matrix(y_test_c, y_pred_gnb)
class_report_gnb = classification_report(y_test_c, y_pred_gnb)

# Display the evaluation metrics for Gaussian Naive Bayes
print("Gaussian Naive Bayes Accuracy:", accuracy_gnb)
print("Gaussian Naive Bayes Confusion Matrix:\n", conf_matrix_gnb)
print("Gaussian Naive Bayes Classification Report:\n", class_report_gnb)
```

```python
from sklearn.model_selection import GridSearchCV
models = {
    'Random Forest': (RandomForestClassifier(), {'n_estimators': [50, 100, 200]}),
    'SVM': (SVC(), {'C': [0.1, 1, 10], 'kernel': ['linear', 'rbf']}),
    'KNN': (KNeighborsClassifier(), {'n_neighbors': [3, 5, 7]}),
    'Gaussian Naive Bayes': (GaussianNB(), {}),
    'XGBoost': (XGBClassifier(), {'n_estimators': [50, 100, 200], 'learning_rate': [0.01, 0.1, 0.2]})
}

# Perform grid search and evaluate each model
for model_name, (model, param_grid) in models.items():
    grid_search = GridSearchCV(model, param_grid, cv=3, scoring='accuracy', n_jobs=-1)
    grid_search.fit(x_train_c, y_train_c)

    # Get the best model from the grid search
    best_model = grid_search.best_estimator_

    # Make predictions on the test data
    y_pred = best_model.predict(x_test_c)

    # Evaluate the model
    accuracy = accuracy_score(y_test_c, y_pred)
    conf_matrix = confusion_matrix(y_test_c, y_pred)
    class_report = classification_report(y_test_c, y_pred)

    # Display the evaluation metrics for each model
    print(f"\n{model_name} - Best Hyperparameters: {grid_search.best_params_}")
    print("Accuracy:", accuracy)
    print("Confusion Matrix:\n", conf_matrix)
    print("Classification Report:\n", class_report)
```
